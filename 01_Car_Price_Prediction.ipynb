```python
def cat_boxplot(df,feature, target):
    df1 = pd.DataFrame(df.groupby([feature])[target].mean().sort_values(ascending = False))
    df1.plot.bar(figsize=(8,6))
    plt.title(f'{feature.upper()} vs Average Price')
    plt.show()
```

```python
cat_features = ['enginetype','CompanyName','fueltype', 'carbody']
target='price'
for feature in cat_features:
    cat_boxplot(df,feature,target)
```

# some others data box and count plot

```python

def plot_count(df, feature,fig):
    plt.figure(figsize=(15,25))
    plt.subplot(6,2,fig)
    fig+=1
    plt.title(feature+' Histogram')
    sns.countplot(df[feature],palette=("magma"))
    plt.subplot(6,2,fig)
    fig+=1
    plt.title(feature+' vs Price')
    sns.boxplot(x=df[feature], y=df.price, palette=("magma"))
plt.show()
    
```

```python

cat_features = ['enginelocation','cylindernumber','fuelsystem', 'drivewheel','doornumber','aspiration']
c=1
for feature in cat_features:
    plot_count(df,feature,c)
```

# Data set of important features

```python
cars = df[['fueltype', 'aspiration','carbody', 'drivewheel','wheelbase','curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 'carlength','carwidth', 'fueleconomy','price']]
```

```python
cars.info()
```

```python
cars['curbweight'] = cars['curbweight'].astype('float')
```

```python
cars['enginesize'] = cars['enginesize'].astype('float')
```

```python
cars['horsepower'] = cars['horsepower'].astype('float')
```

```python
for f in cars.columns:
    if cars[f].dtype!=np.number:
        print(f)
```

# convert all categorical values to dummy variable

```python
for f in cars.columns:
    if cars[f].dtype!=np.number:
        temp = pd.get_dummies(cars[f], drop_first = True)
        cars = pd.concat([cars, temp], axis = 1)
        cars.drop([f], axis = 1, inplace = True)
```

```python
cars.head().T
```

```python
cars.shape
```

```python
X=cars.drop('price', axis=1)
X.head().T
```

```python
y=cars['price']
```

# Select Best Feature

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
feature_select = SelectKBest(score_func=f_classif)
```

```python
feature_select.fit(X,y)
```

```python
score_col = pd.DataFrame(feature_select.scores_, columns=['score_val'])
```

```python
name_col = pd.DataFrame(X.columns, columns=['Name'])
```

```python
top_features = pd.concat([name_col, score_col], axis=1)
top_features
```

```python
top_features.nlargest(14, 'score_val')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=.25, random_state=42)
```
